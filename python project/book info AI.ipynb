```python
import torch ## importing torch and gpt2 , these are the libraries im going to use 
from transformers import GPT2Tokenizer, GPT2LMHeadModel

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
def generate_story(prompt):
    # Encode the prompt
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generate text
    output = model.generate(
        input_ids=input_ids,
        max_length=500,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        top_k=0,
        repetition_penalty=1.2,
        num_return_sequences=1,
    )

    # Decode the output
    story = tokenizer.decode(output[0], skip_special_tokens=True)
    
    return story
```

```python
## this part prompts the user to eneter a name of the story , if the story is not found in gpt2 it will break the loop or if the user writes'exit' 
## for example you can say " Write a story about a time traveler who goes back in time to prevent a historical disaster, but ends up causing it instead." and it will give you stories based on that 
while True:
    prompt = input("Enter a story prompt: ")
    if prompt == "exit":
        break
    story = generate_story(prompt)
    print(story)
```
